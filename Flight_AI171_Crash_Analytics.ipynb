# **🛫 Casualty Metrics**

```python
plt.figure(figsize=(10, 6))
categories = ["Total Onboard", "Passengers", "Crew", "Confirmed Deaths", "Survivors", "Reported Death Toll"]
values = [ df_common["Total_People_Onboard"][0], 
          df_common["Passengers"][0], 
          df_common["Crew"][0], 
          df_common["Confirmed_Deaths"][0], 
          df_common["Survivors"][0], 
          df_common["Death_Toll_Reported"][0]]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b"] 
# Distinct colors
plt.bar(categories, values, color=colors, edgecolor="black")
plt.title("Casualty Metrics for Air India Flight AI171 Crash")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("casualty_metrics.png") 
plt.show()
```

# **👨‍✈️Pilot Experience Comparison**

```python
plt.figure(figsize=(8, 6))

categories = ["Captain Flight Hours", "Co-Pilot Flight Hours", "Required Hours for Commander"]
values = [ df_common["Captain_Flight_Hours"][0], 
           df_common["Co_Pilot_Flight_Hours"][0], 
           df_common["Required_Hours_for_Commander"][0]]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]

# Store bars
bars = plt.bar(categories, values, color=colors, edgecolor="black")

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 10,  
             f'{height:.0f}', ha='center', va='bottom', fontsize=10)
    
plt.title("Pilot Experience for Air India Flight AI171")
plt.xlabel("Category")
plt.ylabel("Flight Hours")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("pilot_experience.png") 
plt.show()
```

# **🚨Survivor vs Fatalities**

```python
plt.figure(figsize=(6, 6))
labels = ["Confirmed Deaths", "Survivors"]
values = [df_common["Confirmed_Deaths"][0], 
          df_common["Survivors"][0]]
colors = ["#d62728", "#9467bd"] 
explode = (0, 0.2) 
plt.pie(values, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90, shadow=True, explode=explode)
plt.tight_layout()
plt.savefig("casualty_pie.png")
plt.show()
```

# **🚒 Emergency Response Resource Comparison**

```python
plt.figure(figsize=(10, 6))
categories = ["NDRF Teams", "Assistance Centers", "Airport Helplines", "DNA Control Room Helplines"]
values = [ df_common["NDRF_Teams_Deployed"][0], 
          df_common["Assistance_Centers_Count"][0], 
          df_common["Airport_Helplines_Count"][0], 
          df_common["DNA_Control_Room_Helplines_Count"][0]]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728"]
plt.bar(categories, values, color=colors, edgecolor="black")
plt.title("Emergency Response Resources for Air India Flight AI171 Crash")
plt.xlabel("Resource Type")
plt.ylabel("Count")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.savefig("emergency_resources.png")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
labels = ["Bodies Handed Over", "Remaining Bodies"]
values = [df_common["Bodies_Handed_Over"][0], 
          df_common["Confirmed_Deaths"][0] - df_common["Bodies_Handed_Over"][0]]
colors = ["#9467bd", "#8c564b"]
explode = (0.1, 0)
plt.pie(values, labels=labels, colors=colors, autopct="%1.1f%%", startangle=90, shadow=True, explode=explode)
plt.title("DNA Processing Progress for Air India Flight AI171 Crash")
plt.tight_layout()
plt.savefig("dna_processing_pie.png")
plt.show()
```
